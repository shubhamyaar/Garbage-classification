```python
# 🧠 Garbage Classification - Test Only Notebook
# Load model and predict uploaded garbage image

# Step 1: Upload the model file (.h5)
from google.colab import files
print("📥 Upload the model file: garbage_classifier.h5")
files.upload()

# Step 2: Load the model
from keras.models import load_model
model = load_model('garbage_classifier.h5')
print("✅ Model loaded successfully!")

# Step 3: Define class labels
class_names = ['cardboard', 'glass', 'metal', 'paper', 'plastic', 'trash']

# Step 4: Upload an image to classify
print("📸 Upload an image to predict")
uploaded = files.upload()

# Step 5: Predict
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image

img_path = list(uploaded.keys())[0]
img = Image.open(img_path).convert("RGB").resize((224, 224))
img_array = image.img_to_array(img) / 255.0
img_array = np.expand_dims(img_array, axis=0)

pred = model.predict(img_array)
predicted_class = class_names[np.argmax(pred)]
confidence = np.max(pred) * 100

# Show result
plt.imshow(img)
plt.axis('off')
plt.title(f"Predicted: {predicted_class} ({confidence:.2f}%)")
plt.show()
```
